#### Since the entire data is too large, we'll only be using the top 20% popular movies

```python
m = df['votes'].quantile(0.80)

dataset = df.copy().loc[df['votes'] >= m]
dataset = dataset.reset_index()

dataset.shape
```

```python
dataset.size
```

# Pre-Processing

```python
dataset.info()
```

```python
dataset.isna().sum()
```

```python
dataset.drop(['original_title', 'year', 'date_published', 'country', 
         'production_company', 'votes', 'budget', 'usa_gross_income',
        'worlwide_gross_income', 'reviews_from_users', 'reviews_from_critics',
        'metascore', 'language', 'user_id'], axis=1, inplace=True)
```

```python
dataset
```

```python
dataset.isna().sum()
```

```python
dataset = dataset.fillna('')
dataset.isna().sum()
```

```python
dataset.info()
```

```python
features = ['director', 'writer', 'actors', 'genre', 'description']
features
```

```python
dataset[features] = dataset[features].replace({'\$': '', ',': '', '\©': '', '¶': '', 'ø': ''}, regex=True)
```

```python
dataset[features]
```

# Implement TF-IDF Vectorizer

```python
dataset['text'] = dataset['director'] + " " + dataset['writer'] + " " + dataset['actors'] + " " + dataset['genre'] + " " + dataset['description']
dataset['text']
```

```python
dataset['text'].iloc[6]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf = TfidfVectorizer(analyzer='word', stop_words='english')
tfidf_matrix = tf.fit_transform(dataset['text'])
tfidf_matrix.shape
```

```python
print(tfidf_matrix)
```

# Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfidf_matrix = tfidf_matrix.astype(np.float32)
cosine_sim = cosine_similarity(tfidf_matrix)
```

```python
print(cosine_sim)
```

### Building a 1-dimensional array with movie titles

```python
titles = dataset['title']
indices = pd.Series(dataset.index, index=dataset['title'])
print(indices.tail(50))
```

# Get Recommendation

```python
new=[]
nd=dict()

def Content_Based_Recommendations(title, dataset):
    
    sim_scores = list(enumerate(cosine_sim[indices[title]]))
    
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    sim_scores = sim_scores[1:100]
    
    movie_indices = [i[0] for i in sim_scores]
    
    new=dataset.iloc[movie_indices]['ratings'].tolist()
    
    j=0
    for i in range(len(new)):
        if new[i]>7:
            nd[j]=dataset.iloc[movie_indices[i],list(range(1, 4))+[-2]]
            j=j+1
    
    return pd.DataFrame.from_dict(nd).transpose()

        
```

```python
result = Content_Based_Recommendations('The Butterfly Ball', dataset)
result = result.sort_values(by='ratings', ascending = False)
```
