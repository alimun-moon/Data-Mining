```python
dataset = [['milk', 'eggs', 'bread'],
           ['milk', 'bread'],
           ['eggs', 'sugar'],
           ['coffee', 'sugar'],
           ['onion', 'coffee', 'carrot', 'icecream']]
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
encoder = TransactionEncoder().fit(dataset)
```

```python
onehot = encoder.transform(dataset)
```

```python
onehot = pd.DataFrame(onehot, columns = encoder.columns_)
```

```python
print(onehot)
```

```python
import scipy.spatial
```

```python
jaccard = scipy.spatial.distance.cdist(onehot, onehot, metric = 'jaccard')
jaccard
```

```python
jaccarddis = pd.DataFrame(jaccard,columns = onehot.index.values,index = onehot.index.values)
```

```python
jaccarddis
```

As we know the less the distance the more similarity we will find. So here 1 is closest to 0, 2 close to 3 and  3 close to 4.

```python
from scipy.spatial import distance
from scipy.spatial.distance import cityblock
```

```python
print(distance.cityblock(jaccarddis[0], jaccarddis[1]))
print(distance.cityblock(jaccarddis[0], jaccarddis[2]))
print(distance.cityblock(jaccarddis[0], jaccarddis[3]))
print(distance.cityblock(jaccarddis[0], jaccarddis[4]))
```

Here, 1 is close to 0

```python
print(distance.cityblock(jaccarddis[1], jaccarddis[2]))
print(distance.cityblock(jaccarddis[1], jaccarddis[3]))
print(distance.cityblock(jaccarddis[1], jaccarddis[4]))


print(distance.cityblock(jaccarddis[2], jaccarddis[3]))
print(distance.cityblock(jaccarddis[2], jaccarddis[4]))

print(distance.cityblock(jaccarddis[3], jaccarddis[4]))
```

2 close to 3 and 4 close to 3
